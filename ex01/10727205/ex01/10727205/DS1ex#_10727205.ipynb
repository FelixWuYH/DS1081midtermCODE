```python
def compare ( count ,i ) :  # 處理此數字裡頭是否有相同數字
    stri = str(count)
    lengh = len(stri)
    run = 0
    cannot = 0
    
    zero = 0   #這裡設代表，假設進來的count 遇到個數字時得 +1
    one = 0
    two = 0 
    three = 0
    four = 0
    five = 0
    six = 0
    seven = 0
    eight = 0
    nine = 0
    
    
    if lengh != i :
        cannot = 1 # 因為數字小於輸入進來的大小，最前面必須補0。ex: i = 3 時，不會有 010 020之類的狀況發生
    
    while ( run < lengh ) :  #從頭開始看，遇到任何數就把那個數的代表+1
        if stri[run] == '0' :
            zero += 1
        if stri[run] == '1' :
            one += 1
        if stri[run] == '2' :
            two += 1
        if stri[run] == '3' :
            three += 1
        if stri[run] == '4' :
            four += 1
        if stri[run] == '5' :
            five += 1
        if stri[run] == '6' :
            six += 1
        if stri[run] == '7' :
            seven += 1
        if stri[run] == '8' :
            eight += 1
        if stri[run] == '9' :
            nine += 1
            
        run += 1
        
    if ( cannot == 1 and zero != 0 ) or ( zero > 1 or one > 1 or two > 1 or three > 1 or four > 1 or five > 1 or six > 1 or seven > 1 or eight > 1 or nine > 1 ) :
        return 0 # 如果有任何數遇到超過一次 就掰掰，下一位 
    else :
        return 1  # 此return代表沒有重複
```
