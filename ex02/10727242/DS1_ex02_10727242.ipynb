找前往終點的路徑，找到一個後就停

```python
def findPath(px, py, width, high, gp):
    
    if map[px][py] == 'E':

        map[px][py] = 'V'

    elif map[px][py] == 'G':

        gp[0] = px
        gp[1] = py
        gp[2] = True

        None

    go = False

    if px - 1 >= 0 and map[px - 1][py] == 'G' and gp[2] == False:
        go = True

        findPath(px - 1, py, width, high, gp)

    if py - 1 >= 0 and map[px][py - 1] == 'G' and gp[2] == False:
        go = True

        findPath(px, py - 1, width, high, gp)

    if py + 1 < high and map[px][py + 1] == 'G' and gp[2] == False:
        go = True

        findPath(px, py + 1, width, high, gp)

    if px + 1 < width and map[px + 1][py] == 'G' and gp[2] == False:
        go = True

        findPath(px + 1, py, width, high, gp)

    if px + 1 < width and (map[px + 1][py] == 'E' or map[px + 1][py] == 'G') and gp[2] == False:
        go = True

        findPath(px + 1, py, width, high, gp)

    if py + 1 < high and (map[px][py + 1] == 'E' or map[px][py + 1] == 'G') and gp[2] == False:
        go = True

        findPath(px, py + 1, width, high, gp)

    if py - 1 >= 0 and (map[px][py - 1] == 'E' or map[px][py - 1] == 'G') and gp[2] == False:
        go = True

        findPath(px, py - 1, width, high, gp)

    if px - 1 >= 0 and (map[px - 1][py] == 'E' or map[px - 1][py] == 'G') and gp[2] == False:
        go = True

        findPath(px - 1, py, width, high, gp)

    if go == False and map[px][py] != 'G':
        map[px][py] = 'V'

        None
```

找終點往起點的最短路徑

```python
def findReadPath(gx, gy, width, high, c):

    if c[0] == False and map[gx][gy] == 'V':
        map[gx][gy] = 'R'

    if gx == 0 and gy == 0:
        map[gx][gy] = 'R'

        c[0] = True

        None

    if c[0] == False:

        if gy - 1 >= 0 and map[gx][gy - 1] == 'V':
            findReadPath(gx, gy - 1, width, high, c)

        if gx - 1 >= 0 and map[gx - 1][gy] == 'V':
            findReadPath(gx - 1, gy, width, high, c)

        if gx + 1 < width and map[gx + 1][gy] == 'V':
            findReadPath(gx + 1, gy, width, high, c)

        if gy + 1 < high and map[gx][gy + 1] == 'V':
            findReadPath(gx, gy + 1, width, high, c)
```

印出最終答案(真實路徑)

```python
def printAns(map):

    for i in range(0, high):

        for j in range(0, width):

            if map[j][i] == 'V':
                map[j][i] = 'E'

    for i in range(0, high):

        for j in range(0, width):
            print(map[j][i], end='')

        print()
```

主程式(讀檔->找路徑->找最短路徑->印答案)

```python
fileName = input("輸入檔案名稱(輸入0離開)：")

while fileName is not "0" :

    isOpen = True

    try:

        f = open(fileName, 'r')

    except FileNotFoundError as n:

        isOpen = False

        print( fileName + "不存在請重新輸入!!")
        print()

    if isOpen:

        r = f.readline()

        g = r.split(' ', 2)[0]

        width = int(g)

        g = r.split(' ', 2)[1].split('\n', 2)[0]

        high = int(g)

        map = [[0]*high for i in range(width)]

        for i in range(0, high) :

            for j in range(0, width) :

                s = f.read(1)

                while  s == '\r' or s == '\n' :

                    s = f.read(1)

                map[j][i] = s

        gp = [-1,-1, False]

        findPath( 0, 0, width, high, gp )

        for i in range(0, high) :

            for j in range(0, width) :

                print(map[j][i], end='')

            print()

        if gp[2] == True:

            c = [False]
            findReadPath(gp[0], gp[1], width, high, c)
            print()
            printAns(map)

    print()
    fileName = input("輸入檔案名稱(輸入0離開)：")
```
